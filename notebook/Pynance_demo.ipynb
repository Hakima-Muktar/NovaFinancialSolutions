```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt       

```

#Load News

```python

file_path = "data/news/raw_analyst_ratings.csv" 
if not os.path.exists(file_path):
    print(f"File not found: {file_path}")
else:
    df = pd.read_csv(file_path)
    print("Columns in CSV:", df.columns.tolist())
    first_col = df.columns[0]
    try:
        df[first_col] = pd.to_datetime(df[first_col])
        print(f"Column '{first_col}' converted to datetime successfully.")
    except Exception as e:
        print(f"Could not convert '{first_col}' to datetime:", e)
    print("Data preview:")
    print(df.head())
```

# Headline Length Statistics

```python
from news import headline_length_stats
stats = headline_length_stats(df)
print("Headline Length Statistics:")
print(stats)
plt.figure(figsize=(8,4))
df["headline"].astype(str).str.len().hist(bins=40)
plt.title("Headline Length Distribution")
plt.xlabel("Characters")
plt.ylabel("Count")
plt.show()
```
