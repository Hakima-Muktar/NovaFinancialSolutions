```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt

```

#Load News

```python
# load the dataset
df = pd.read_csv('../data/news/raw_analyst_ratings.csv')
```

```python
# heading of the dataset
print (df.head())
```

```python
# statistical summary
print(df.describe())
```

```python
# missing values
print(df.isnull().sum())
```

```python
# duplicated values
print(df.duplicated().sum())
```

```python
# count length
df['headline_length']=df['headline'].apply(len)
```

```python
# time series plot of stock over date
plt.figure(figsize=(10,6))
df["stock"].astype(str).str.len().hist(bins=20)
plt.title("time series plot of stock over date")
# Setting the X and Y labels
plt.xlabel('stock')
plt.ylabel('date')
# Adding the legends
plt.show()
```

```python
plt.figure(figsize=(8,4))
df["headline"].astype(str).str.len().hist(bins=30)
plt.title("Headline Length Distribution")
plt.xlabel("Characters")
plt.ylabel("Count")
plt.show()
```

```python
# visulize distribution of target publisher variable
plt.figure(figsize=(10,6))
plt.bar(df['publisher'].value_counts().head(10).index, df['publisher'].value_counts().head(10).values)
plt.title("Top 10 Publishers")
plt.xlabel("Publisher")
plt.ylabel("Articles")
plt.xticks(rotation=45)
plt.show()
```

```python
# visulize distribution of target variable
plt.figure(figsize=(10,6))
df["date"].astype(str).str.len().hist(bins=30)
plt.title("Distribution of headlines by date")
plt.xlabel('date')
plt.ylabel('headlines')
plt.show()
```
